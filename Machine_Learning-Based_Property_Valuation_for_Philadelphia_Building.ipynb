# **Business Problem**

**Context**

The Office of Property Assessment (OPA) is a department within the City of Philadelphia government responsible for assessing the value of all properties located within the city. The property values determined by OPA serve as an important reference for property owners and buyers when establishing transaction prices. In addition, these assessed values form the basis for calculating property taxes that property owners are required to pay. Property tax is one of the primary sources of revenue used to support the funding of public schools in the City of Philadelphia.

As the authority responsible for property assessment, OPA is committed to producing property valuations that are fair, accurate, and easily understood by the public. This ensures that properties with similar characteristics and market values are assessed consistently and are subject to equitable taxation. In determining a property's value, OPA considers a variety of factors, including property sales data and the physical and market characteristics of the property.

Property valuation must be conducted carefully and reflect the actual condition of each property. If a property's assessed value is set below its true market value, government revenue from property taxes may decline, potentially affecting the funding available for public schools in Philadelphia. In addition, property owners may raise concerns because assessed values can influence the market perception and transaction prices of their properties. Conversely, if property values are assessed too high, property owners may file complaints due to the increased property tax burden resulting from the higher assessment.

**Problem Statement**

The Office of Property Assessment (OPA) is responsible for assessing the value of all properties in the City of Philadelphia to ensure fair taxation and provide reliable property valuation information to the public. Given the large number of properties and the diversity of their characteristics, determining accurate property values is a complex and time-consuming process. Property assessments require extensive analysis of historical sales transactions, market trends, neighborhood conditions, and individual property attributes such as size, age, location, building condition, and other relevant factors.

Currently, conducting a comprehensive property valuation can take anywhere from three months to one year, depending on the complexity of the analysis and the availability of supporting data. This lengthy assessment process creates challenges for OPA in maintaining up-to-date property valuations and responding efficiently to changes in the real estate market. Delays or inaccuracies in property assessments may result in unfair taxation, reduced public trust, and potential impacts on property tax revenue that supports public services, including the funding of Philadelphia's public schools.

To address these challenges, OPA requires a tool that can support and accelerate the property valuation process while maintaining a high level of accuracy and consistency. Specifically, OPA needs a Machine Learning-based Property Valuation Tool capable of analyzing historical property sales data and property characteristics to predict market values. By leveraging machine learning models, the tool can identify complex relationships between property features and market prices, enabling OPA to generate more accurate valuation estimates in a significantly shorter time frame. **Analytical Approach**

This study employs a regression-based analytical approach to predict the market value of properties in Philadelphia. The methodology consists of data preprocessing, including data cleansing, feature selection, and variable encoding, followed by model development and evaluation. The resulting model is analyzed to identify the primary drivers of property values, providing actionable insights that can be used to formulate recommendations for OPA and support a more accurate, consistent, and data-driven property valuation process.

**Metric Evaluation**

To select the most appropriate model for predicting property values, it is essential to use suitable evaluation metrics. In this study, we employ Mean Absolute Percentage Error (MAPE) and R-squared (R²) as the primary evaluation metrics.

MAPE is chosen because the dataset contains a significant number of outliers and a wide range of property values. This is expected, as the dataset includes various types of properties, ranging from residential properties to specialized properties such as stadiums.

MAPE measures the prediction error by calculating the percentage difference between the predicted value and the actual value. Specifically, it is computed as the absolute value of the residual (error) divided by the actual value and multiplied by 100 percent.

The interpretation of MAPE values is as follows:

| MAPE Value	| Interpretation |
|---|---|
| 0% – 10% |	Excellent |
| 10% – 20%	| Good |
| 20% – 50% |	Reasonable |
| > 50% |	Inaccurate |

In addition to MAPE, R-squared (R²) is used to evaluate the explanatory power of the model. R-squared measures the proportion of variance in the dependent variable that can be explained by the independent variables. The value of R-squared ranges from 0 to 1, where a higher value indicates that the independent variables collectively explain a larger proportion of the variation in the dependent variable.

Therefore, the R-squared statistic is used to assess the extent to which the selected independent variables contribute to explaining variations in property values, providing an indication of the model’s overall goodness of fit.

There are three categories commonly used to interpret R-squared values: strong, moderate, and weak. An R-squared value of 0.75 is considered strong, an R-squared value of 0.50 is considered moderate, and an R-squared value of 0.25 is considered weak.

## 2.3 Outlier Treatment

### 2.3.1  Structural Column Outliers

Per the project requirements, the following columns are inspected for values that are **physically impossible or clearly data-entry errors**. Rows containing such values are **dropped** since they represent corrupted records that would distort aggregations.

| Column | Reasonable Range | Rationale |
|--------|-----------------|-----------|
| `year_built` | 1650 – 2026 | Philadelphia oldest building ~1652; 0 = unknown |
| `number_of_bedrooms` | 0 – 20 | 50-bedroom "residential" is impossible; likely data error |
| `number_of_bathrooms` | 0 – 12 | Rare luxury homes may have 10+; >12 is implausible |
| `number_stories` | 0 – 15 | Philadelphia tallest residential ~10 stories |
| `garage_spaces` | 0 – 20 | Keeps small commercial parking lots |

```python
df["year_built"] = df["year_built"].replace(0, np.nan)

outlier_cols = [
    ("year_built",          1650, 2026,  "Year Built"),
    ("number_of_bedrooms",  0,    20,    "Bedrooms"),
    ("number_of_bathrooms", 0,    12,    "Bathrooms"),
    ("number_stories",      0,    15,    "Stories"),
    ("garage_spaces",       0,    20,    "Garage Spaces")
]

# Remove outlier rows (with category exceptions)

n_before = len(df)
drop_mask = pd.Series(False, index=df.index)

allowed_categories = ["Commercial", "Industrial"]

report_rows = []

for col, lo, hi, label in outlier_cols:

    # default rule
    mask = (df[col].notna()) & ((df[col] < lo) | (df[col] > hi))

    # exception for stories
    if col == "number_stories":
        mask = (
            (df[col].notna())
            & ((df[col] < lo) | (df[col] > hi))
            & (~df["category_code_description"].isin(allowed_categories))
        )

    # exception for garage spaces
    elif col == "garage_spaces":
        mask = (
            (df[col].notna())
            & ((df[col] < lo) | (df[col] > hi))
            & (~df["category_code_description"].isin(allowed_categories))
        )

    report_rows.append({
        "Column": col,
        "Lower Bound": lo,
        "Upper Bound": hi,
        "Outlier Rows": int(mask.sum()),
        "Outlier %": f"{mask.mean()*100:.3f}%"
    })

    drop_mask |= mask

df = df[~drop_mask].copy()

n_dropped = n_before - len(df)

print(pd.DataFrame(report_rows).to_string(index=False))
print(f"Total rows dropped (union of all outlier masks) : {n_dropped:,}")
print(f"Dataset size after structural outlier removal    : {len(df):,} rows")
```

## 2.4 Missing Value Handling

Before applying any imputation we identify each column's **missingness type** to choose the most defensible strategy.

| Missingness Type | Description | Treatment |
| ---------- | ---------------------- | ---------------- |
| **Structural Missingness** | Missing values represent a meaningful absence rather than lost information. | Use indicator variables and appropriate placeholder values (e.g., 0, `"None"`). |
| **Property-Type Dependent Missingness** | Missing values are associated with differences in property categories and land uses. | Impute using category-specific median or mode. |
| **Minor Random Missingness** | Small amounts of missing data without a clear structural explanation. | Impute using global median or mode, or drop rows |

```python
# Current null counts
null_df = pd.DataFrame({
    "null_count": df.isnull().sum(),
    "null_pct"  : (df.isnull().sum() / len(df) * 100).round(2),
    "dtype"     : df.dtypes,
}).query("null_count > 0").sort_values("null_pct", ascending=False)

print(f"Dataset rows           : {len(df):,}")
print(f"Total null cells       : {df.isnull().sum().sum():,}")
print(f"Columns with NaN       : {len(null_df)}")
print()
print(null_df[["null_count", "null_pct", "dtype"]].to_string())
```

```python
print("=" * 80)
print("MISSING VALUE IMPUTATION")
print("=" * 80)

n_initial = len(df)

# ----------------------------------------------------------------------------
# 1. CENTRAL AIR
# ----------------------------------------------------------------------------
# Assumption:
# - 0 = No Central Air ("N")
# - Vacant Land (6) and Industrial (5) without central air information
#   are assumed to have no central air.

df["central_air"] = df["central_air"].replace(0, "N")

mask = (
    df["central_air"].isna()
    & df["category_code"].isin([5, 6])
)
df.loc[mask, "central_air"] = "N"

remaining = df["central_air"].isna().sum()
df["central_air"] = df["central_air"].fillna("Unknown")

print(
    f"central_air               → labelled {remaining:,} rows as 'Unknown'"
)

# ----------------------------------------------------------------------------
# 2. TYPE HEATER
# ----------------------------------------------------------------------------
# Assumption:
# - Vacant Land (6) and Industrial (5) without heating information
#   are assumed to have no heating system.

mask = (
    df["type_heater"].isna()
    & df["category_code"].isin([5, 6])
)
df.loc[mask, "type_heater"] = 0

remaining = df["type_heater"].isna().sum()
df["type_heater"] = df["type_heater"].fillna("Unknown")

print(
    f"type_heater               → labelled {remaining:,} rows as 'Unknown'"
)

# ----------------------------------------------------------------------------
# 3. GARAGE TYPE
# ----------------------------------------------------------------------------
# Assumption:
# - Vacant Land (6) has no garage
# - garage_spaces = 0 implies no garage

mask = (
    df["garage_type"].isna()
    & (df["category_code"] == 6)
)
df.loc[mask, "garage_type"] = 0

mask = (
    df["garage_type"].isna()
    & (df["garage_spaces"] == 0)
)
df.loc[mask, "garage_type"] = 0

remaining = df["garage_type"].isna().sum()
df["garage_type"] = df["garage_type"].fillna("Unknown")

print(
    f"garage_type               → labelled {remaining:,} rows as 'Unknown'"
)

# ----------------------------------------------------------------------------
# 4. BASEMENTS
# ----------------------------------------------------------------------------
# Assumption:
# - Vacant Land (6) has no basement
# - Garage Type A (built-in / basement garage) implies partial basement

mask = (
    df["basements"].isna()
    & (df["category_code"] == 6)
)
df.loc[mask, "basements"] = 0

mask = (
    df["basements"].isna()
    & (df["garage_type"] == "A")
)
df.loc[mask, "basements"] = "H"

remaining = df["basements"].isna().sum()
df["basements"] = df["basements"].fillna("Unknown")

print(
    f"basements                 → labelled {remaining:,} rows as 'Unknown'"
)

# ----------------------------------------------------------------------------
# 5. GENERAL CONSTRUCTION
# ----------------------------------------------------------------------------
# Assumption:
# - Vacant Land (6) has no construction type

mask = (
    df["general_construction"].isna()
    & (df["category_code"] == 6)
)
df.loc[mask, "general_construction"] = 0

remaining = df["general_construction"].isna().sum()
df["general_construction"] = df["general_construction"].fillna("Unknown")

print(
    f"general_construction      → labelled {remaining:,} rows as 'Unknown'"
)

# ----------------------------------------------------------------------------
# 6. YEAR BUILT
# ----------------------------------------------------------------------------
# Critical feature — remove unknown build years

before = len(df)

df = df[
    df["year_built"].notna()
    & (df["year_built"] != 0)
].copy()

print(f"year_built                → dropped {before - len(df):,} rows")

# ----------------------------------------------------------------------------
# 7. MODE IMPUTATION WITHIN CATEGORY
# ----------------------------------------------------------------------------

mode_cols = [
    "interior_condition",
    "exterior_condition",
    "topography",
    "view_type"
]

for col in mode_cols:
    df[col] = (
        df.groupby("category_code")[col]
        .transform(
            lambda x: x.fillna(
                x.mode().iloc[0]
                if not x.mode().empty
                else np.nan
            )
        )
    )

print("category mode imputation  → completed")

# ----------------------------------------------------------------------------
# 8. DROP REMAINING LOW-MISSING FEATURES (<3%)
# ----------------------------------------------------------------------------

low_missing_cols = [
    "geographic_ward",
    "parcel_shape",
    "zoning",
    "sale_date",
    "off_street_open",
    "depth",
    "fireplaces",
    "number_of_bathrooms",
    "frontage",
    "garage_spaces",
    "number_stories",
    "number_of_rooms",
    "number_of_bedrooms",
    "total_area",
    "total_livable_area"
]

before = len(df)

df = df.dropna(subset=low_missing_cols)

print(f"low-missing features      → dropped {before - len(df):,} rows")

# ----------------------------------------------------------------------------
# FINAL SUMMARY
# ----------------------------------------------------------------------------

print("-" * 80)
print(f"Initial rows : {n_initial:,}")
print(f"Final rows   : {len(df):,}")
print(f"Rows removed : {n_initial - len(df):,}")
print(f"Retention    : {(len(df)/n_initial)*100:.2f}%")

remaining_nulls = df.isnull().sum().sum()
print(f"Remaining NaN cells : {remaining_nulls:,}")

if remaining_nulls > 0:
    print("\nColumns still containing NaN:")
    print(df.isnull().sum()[df.isnull().sum() > 0].sort_values(ascending=False))
else:
    print("\n✓ No missing values remain.")

print("-" * 80)
print("\nRemaining columns:")
print(df.columns)
print(f"\nDataset shape: {df.shape[0]:,} rows × {df.shape[1]:,} columns")
```

| 2 | Label as `"Unknown"` | Preserves observations and lets models learn from missingness. |

## 2.5 Feature Engineering

New features are created to enrich both analytical and modelling workflows.

```python
# 1. Building Age
# Rationale: Captures depreciation & modernisation signal. Interpretable in years.
df["building_age"] = np.where(
    df["year_built"].notna(),
    CURRENT_YEAR - df["year_built"],
    np.nan
)
df.loc[df["building_age"] < 0, "building_age"] = np.nan   # built "in the future" = error

# 2. Building Era
era_bins   = [0, 1899, 1949, 1979, 1999, 2100]
era_labels = ["Historic (<1900)", "Early 20th (1900-49)",
              "Mid-century (1950-79)", "Modern (1980-99)", "Contemporary (2000+)"]
df["building_era"] = pd.cut(df["year_built"], bins=era_bins,
                             labels=era_labels, right=True)
df["building_era"] = df["building_era"].cat.add_categories("Unknown").fillna("Unknown")

# 3. Price per Livable Area (Market Value / Livable Area)
# Rationale: Normalises size; the classic $/sqft metric used by all appraisers.
df["price_per_livable_area"] = np.where(
    df["total_livable_area"] > 0,
    df["market_value"] / df["total_livable_area"],
    np.nan
)

# 4. Livable Area Ratio (Livable Area / Total Lot Area)
# Rationale: Measures development intensity; high ratio = dense urban lot.
df["livable_area_ratio"] = np.where(
    df["total_area"] > 0,
    df["total_livable_area"] / df["total_area"],
    np.nan
)
df["livable_area_ratio"] = df["livable_area_ratio"].clip(upper=5)  # cap impossible ratios

# 5. Lot Size (Frontage × Depth)
# Rationale: Alternative area measure; frontage captures street-facing value.
df["lot_size_calc"] = np.where(
    (df["frontage"] > 0) & (df["depth"] > 0),
    df["frontage"] * df["depth"],
    np.nan
)

# 6. Sale Date → Year / Month
df["sale_date"]  = pd.to_datetime(df["sale_date"], errors="coerce")
df["sale_year"]  = df["sale_date"].dt.year.fillna(0).astype(int)
df["sale_month"] = df["sale_date"].dt.month.fillna(0).astype(int)
df = df.drop(columns=["sale_date"])

# 7. Binary Amenity Flags
df["has_central_air"] = (df["central_air"].astype(str).str.upper() == "Y").astype(int)
df["has_garage"]      = (df["garage_spaces"] > 0).astype(int)
df["has_fireplace"]   = (df["fireplaces"] > 0).astype(int)
# basements: D = no basement, 0 = not recorded
df["has_basement"] = (~df["basements"].isin(["D", "0", "0", None]) &
                       df["basements"].notna()).astype(int)

# 8. Market Value Tier
tier_bins = [0, 100_000, 400_000, 800_000, float("inf")]
tier_labels = [
    "Low",
    "Medium",
    "High",
    "Luxury"]
df["value_tier"] = pd.cut(
    df["market_value"],
    bins=tier_bins,
    labels=tier_labels,
    include_lowest=True)

print("Feature engineering complete. New columns added:")
new_feats = ["building_age","building_era","price_per_livable_area","livable_area_ratio",
             "lot_size_calc","sale_year","sale_month",
             "has_central_air","has_garage","has_fireplace","has_basement","value_tier"]
for feat in new_feats:
    desc = df[feat].describe() if df[feat].dtype in ["float64","int64","int32"] else df[feat].value_counts()
    print(f"  {feat:<30}  dtype={str(df[feat].dtype):<12}  nunique={df[feat].nunique()}")
```

### 3.7.2 Correlation Ratio (Eta) - Categorical Associations

```python
from scipy.stats import f_oneway

# Correlation Ratio (Eta)

def correlation_ratio(categories, values):
    """
    Correlation Ratio (Eta)
    categories : categorical variable
    values     : continuous variable
    """

    categories = pd.Series(categories)
    values = pd.Series(values)

    valid = categories.notna() & values.notna()

    categories = categories[valid]
    values = values[valid]

    grand_mean = values.mean()

    ss_between = 0
    ss_total = ((values - grand_mean) ** 2).sum()

    for cat in categories.unique():
        group = values[categories == cat]
        ss_between += len(group) * (group.mean() - grand_mean) ** 2

    eta = np.sqrt(ss_between / ss_total)
    return eta

cat_cols = [
    "category_code_description",
    "building_code_description",
    "general_construction",
    "zoning",
    "topography",
    "view_type",
    "type_heater",
    "basements",
    "parcel_shape",
    "building_era",
]

cat_cols = [c for c in cat_cols if c in df_viz.columns]

eta_scores = {}

for col in cat_cols:

    eta_scores[col] = correlation_ratio(
        df_viz[col],
        df_viz["market_value"]
    )

eta_series = (
    pd.Series(eta_scores)
      .sort_values()
)

fig, ax = plt.subplots(figsize=(11,6))

bars = ax.barh(
    eta_series.index,
    eta_series.values,
    color=C["teal"],
    alpha=.85,
    edgecolor="white"
)

ax.set_xlabel("Correlation Ratio (η)")
ax.set_title(
    "Correlation Ratio (η)\nCategorical Features vs Market Value",
    fontweight="bold"
)

for bar,val in zip(bars,eta_series.values):

    ax.text(
        val+0.01,
        bar.get_y()+bar.get_height()/2,
        f"{val:.3f}",
        va="center"
    )

plt.tight_layout()
plt.show()

print(eta_series.sort_values(ascending=False))
```

**Feature Association Analysis**

Spearman's Rank Correlation is used to evaluate the relationship between numerical/ordinal features and the target `market_value` because this method is able to capture monotonic relationships without assuming a linear relationship and is more robust against outliers. For categorical features, the Correlation Ratio (η) is used because this method specifically measures the strength of the association between categorical variables and continuous variables. Unlike Cramér's V, which requires both variables to be categorical so that `market_value` must be discretized first, the Correlation Ratio can utilize continuous `market_value` values ​​directly. Thus, all numerical information on the target is retained so that the analysis results are more in line with the research objective, namely to build a regression model to predict property market value.

## 4.2 Feature Selection & Train/Test Split

### Feature Selection Strategy

Features are selected based on three criteria established in the EDA (Sections 3.7):

1. **Predictive power** — confirmed by Spearman correlation and RF Feature Importance
2. **Business relevance** — meaningful to OPA assessors and defensible in public hearings
3. **No data leakage** — excludes variables OPA would not know *before* assessment
   (e.g. `sale_price`, `price_per_sqft_sale`, `price_per_livable_area`)

**Excluded features and rationale:**

| Feature(s) | Reason for Exclusion |
|---|---|
| `sale_price`, `price_per_sqft_sale`, `price_per_livable_area` | Leakage: derived from or correlated with `market_value` post-sale |

| `year_built` | Replaced by `building_age` and `building_era` |
| `lot_size_calc`, `value_tier` | Derived from the target; present in `df_ml` for EDA, excluded for ML |
| `building_code` | >800 categories — high cardinality noise with minimal signal gain |

### Train/Test Split Strategy
- **Ratio**: 80 % train / 20 % test
- **Stratification**: on `category_code` to maintain equal category representation across splits
- **Rationale**: With ~520K rows, 20% gives ~104K test samples — sufficient for stable metric estimation

## 4.3 Baseline Model Comparison

Seven algorithms spanning the full complexity spectrum are evaluated as baselines.
Every model is wrapped in an identical `sklearn.Pipeline` that handles:

- Median imputation → Standard Scaling for numerical features
- Constant imputation ("Unknown") → One-Hot Encoding for low-cardinality categoricals
- Constant imputation ("Unknown") → Ordinal Encoding for high-cardinality / ordered categoricals

This ensures a fair, leakage-free comparison where each model receives identically preprocessed data.

| Model | Type | Strengths | Weaknesses |
|---|---|---|---|
| **Linear Regression** | Linear | Interpretable, fast | Assumes linearity — poor for property data |
| **Ridge Regression** | Linear + L2 | Handles multicollinearity | Still linear |
| **Lasso Regression** | Linear + L1 | Automatic feature selection | Can drop important features |
| **Extra Trees** | Randomised Ensemble | Low variance, fast | Can be less precise than RF |
| **Random Forest** | Bagged Ensemble | Robust, handles non-linearity | Slow, memory-intensive |
| **Gradient Boosting** | Sequential Boosting | High accuracy | Overfitting risk if untuned |
| **XGBoost** | Optimised Boosting | Fastest among boosters, regularised | Many hyperparameters |


```python
def to_string_array(x):
    """Cast a column (possibly mixed int/str, e.g. 0 mixed with 'A'/'D' codes)
    to a uniform string dtype before one-hot/ordinal encoding."""
    return x.astype(str)

# ── Preprocessing pipeline ────────────────────────────────────────────────────

num_pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler",  StandardScaler()),
])

cat_ohe_pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="constant", fill_value="Unknown")),
    ("to_string",
     FunctionTransformer(
         to_string_array,
         validate=False
     )),
    ("encoder",
     OneHotEncoder(
         handle_unknown="ignore",
         sparse_output=False
     )),
])

cat_ord_pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="constant", fill_value="Unknown")),
    ("to_string",
     FunctionTransformer(
         to_string_array,
         validate=False
     )),
    ("encoder",
     OrdinalEncoder(
         handle_unknown="use_encoded_value",
         unknown_value=-1
     )),
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num",     num_pipe,     NUM_FEATURES),
        ("cat_ohe", cat_ohe_pipe, CAT_FEATURES_OHE),
        ("cat_ord", cat_ord_pipe, CAT_FEATURES_ORD),
    ],
    remainder="drop",
)

print("✅  Preprocessor pipeline built.")
print(f"   Numerical features   : {len(NUM_FEATURES)}")
print(f"   OHE cat features     : {len(CAT_FEATURES_OHE)}")
print(f"   Ordinal cat features : {len(CAT_FEATURES_ORD)}")
```

```python
# ── Define baseline model pipelines ──────────────────────────────────────────

BASELINE_MODELS = {
    "Linear Regression": Pipeline([
        ("prep",  preprocessor),
        ("model", LinearRegression()),
    ]),
    "Ridge Regression": Pipeline([
        ("prep",  preprocessor),
        ("model", Ridge(alpha=1.0, random_state=RANDOM_STATE)),
    ]),
    "Lasso Regression": Pipeline([
        ("prep",  preprocessor),
        ("model", Lasso(alpha=0.001, max_iter=5000, random_state=RANDOM_STATE)),
    ]),
    "Extra Trees": Pipeline([
        ("prep",  preprocessor),
        ("model", ExtraTreesRegressor(
            n_estimators=200, max_depth=20, min_samples_leaf=3,
            max_features="sqrt", n_jobs=-1, random_state=RANDOM_STATE,
        )),
    ]),
    "Random Forest": Pipeline([
        ("prep",  preprocessor),
        ("model", RandomForestRegressor(
            n_estimators=200, max_depth=20, min_samples_leaf=3,
            max_features="sqrt", n_jobs=-1, random_state=RANDOM_STATE,
            oob_score=True,
        )),
    ]),
    "Gradient Boosting": Pipeline([
        ("prep",  preprocessor),
        ("model", GradientBoostingRegressor(
            n_estimators=300, learning_rate=0.05, max_depth=6,
            subsample=0.8, min_samples_leaf=10, random_state=RANDOM_STATE,
        )),
    ]),
    "XGBoost" : Pipeline([
        ("prep",  preprocessor),
        ("model", xgb.XGBRegressor(
            n_estimators=400, learning_rate=0.05, max_depth=6,
            subsample=0.8, colsample_bytree=0.8, min_child_weight=5,
            reg_alpha=0.1, reg_lambda=1.0, tree_method="hist",
            eval_metric="rmse", random_state=RANDOM_STATE, n_jobs=-1, verbosity=0,
        )),
    ]),
}

print(f"✅  {len(BASELINE_MODELS)} baseline model pipelines created:")
for name in BASELINE_MODELS:
    print(f"   • {name}")
```

```python
# ── Metric helpers ────────────────────────────────────────────────────────────

def mape_score(y_true, y_pred, is_log=True):
    # MAPE computed in original USD scale (not log space)
    yt = np.expm1(y_true) if is_log else np.asarray(y_true)
    yp = np.expm1(y_pred) if is_log else np.asarray(y_pred)
    mask = yt > 0
    return np.mean(np.abs((yt[mask] - yp[mask]) / yt[mask])) * 100


def evaluate_model(name, pipeline, X_tr, y_tr, X_te, y_te):
    t0 = time.time()
    pipeline.fit(X_tr, y_tr)
    elapsed = time.time() - t0

    y_tr_pred = pipeline.predict(X_tr)
    y_te_pred = pipeline.predict(X_te)

    return {
        "Model"           : name,
        "Test R²"         : r2_score(y_te, y_te_pred),
        "Test MAPE (%)"   : mape_score(y_te, y_te_pred),
        "Test RMSE (log)" : np.sqrt(mean_squared_error(y_te, y_te_pred)),
        "Test MAE ($)"    : mean_absolute_error(np.expm1(y_te), np.expm1(y_te_pred)),
        "Train R²"        : r2_score(y_tr, y_tr_pred),
        "Train MAPE (%)"  : mape_score(y_tr, y_tr_pred),
        "Overfit ΔR²"     : r2_score(y_tr, y_tr_pred) - r2_score(y_te, y_te_pred),
        "Train Time (s)"  : round(elapsed, 1),
    }, pipeline


# ── Train & evaluate all baselines ───────────────────────────────────────────
print("Training baseline models…\n")
results_list   = []
trained_models = {}

for name, pipe in BASELINE_MODELS.items():
    print(f"  [{name}] training…")
    metrics, fitted = evaluate_model(name, pipe, X_train, y_train, X_test, y_test)
    results_list.append(metrics)
    trained_models[name] = fitted
    print(f"    Test R²   : {metrics['Test R²']:.4f}")
    print(f"    Test MAPE : {metrics['Test MAPE (%)']:.2f} %")
    print(f"    Train R²  : {metrics['Train R²']:.4f}")
    print(f"    Time      : {metrics['Train Time (s)']} s\n")

df_results = (
    pd.DataFrame(results_list)
    .sort_values("Test R²", ascending=False)
    .reset_index(drop=True)
)

DISP_COLS = ["Model","Test R²","Test MAPE (%)","Test RMSE (log)",
             "Test MAE ($)","Train R²","Overfit ΔR²","Train Time (s)"]
print("=" * 75)
print("BASELINE MODEL RESULTS — sorted by Test R²")
print("=" * 75)
print(df_results[DISP_COLS].to_string(index=False))
```

## 4.4 Cross-Validation — Top-3 Model Robustness Check

Single train/test splits can be influenced by the random seed. To obtain stable, unbiased
performance estimates we apply **5-Fold Cross-Validation** on the training set for the top-3
baseline models. This provides:

1. Mean performance across 5 independent validation windows
2. Standard deviation — a direct measure of model stability
3. Confidence that the selected model will generalise, not just get lucky on one split

```python
# ── Custom negative-MAPE scorer for cross_val_score ──────────────────────────
def neg_mape_scorer(y_true, y_pred):
    yt = np.expm1(y_true); yp = np.expm1(y_pred)
    mask = yt > 0
    return -np.mean(np.abs((yt[mask] - yp[mask]) / yt[mask])) * 100

mape_scorer = make_scorer(neg_mape_scorer)

# ── Top 3 models by Test R² ───────────────────────────────────────────────────
TOP3 = df_results["Model"].head(3).tolist()
print(f"Top-3 models selected for cross-validation: {TOP3}\n")

kf = KFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
cv_results = {}

for name in TOP3:
    fresh_pipe = BASELINE_MODELS[name]
    print(f"  Running 5-fold CV for [{name}] …")
    t0 = time.time()

    r2_scores = cross_val_score(fresh_pipe, X_train, y_train,
                                 cv=kf, scoring="r2", n_jobs=-1)

    # Subsample 50K for MAPE CV to keep runtime manageable
    rng   = np.random.default_rng(RANDOM_STATE)
    idx_s = rng.choice(len(X_train), size=min(50_000, len(X_train)), replace=False)
    mape_scores = cross_val_score(fresh_pipe,
                                   X_train.iloc[idx_s], y_train.iloc[idx_s],
                                   cv=kf, scoring=mape_scorer, n_jobs=-1)

    cv_results[name] = {
        "CV R² Mean"  : r2_scores.mean(),
        "CV R² Std"   : r2_scores.std(),
        "CV MAPE Mean": -mape_scores.mean(),
        "CV MAPE Std" : mape_scores.std(),
        "Time (s)"    : round(time.time()-t0, 1),
    }
    print(f"    R²   : {r2_scores.mean():.4f} ± {r2_scores.std():.4f}")
    print(f"    MAPE : {-mape_scores.mean():.2f} % ± {mape_scores.std():.2f} %")
    print(f"    Time : {cv_results[name]['Time (s)']} s\n")

df_cv = (
    pd.DataFrame(cv_results).T
    .reset_index().rename(columns={"index": "Model"})
)
print("=" * 65)
print("5-FOLD CROSS-VALIDATION RESULTS")
print("=" * 65)
print(df_cv.to_string(index=False))
```

## 4.5 Hyperparameter Tuning — Randomized Search CV

`RandomizedSearchCV` with 50 iterations and 3-fold inner CV is used to find the optimal
hyperparameter combination. This approach is preferred over exhaustive Grid Search because:

- The hyperparameter space is large (thousands of combinations per grid)
- Random sampling with 50 iterations provides strong coverage at a fraction of the compute cost
- A 200K-row subsample of the training data is used for speed without sacrificing statistical
  representativeness (>35% of the full training set)

The final model is then re-fitted on the **complete training set** (not just the tuning subsample)
to maximise the information available to the model before test-set evaluation.

```python
# ── Parameter grids (conditional on BEST_MODEL_NAME) ─────────────────────────

if "XGBoost" in BEST_MODEL_NAME:
    param_dist = {
        "model__n_estimators"    : [300, 400, 500, 600],
        "model__learning_rate"   : [0.01, 0.03, 0.05, 0.08, 0.1],
        "model__max_depth"       : [4, 5, 6, 7, 8],
        "model__subsample"       : [0.6, 0.7, 0.8, 0.9],
        "model__colsample_bytree": [0.6, 0.7, 0.8, 0.9],
        "model__min_child_weight": [3, 5, 7, 10],
        "model__reg_alpha"       : [0, 0.01, 0.05, 0.1, 0.5],
        "model__reg_lambda"      : [0.5, 1.0, 1.5, 2.0],
    }
elif "Random Forest" in BEST_MODEL_NAME:
    param_dist = {
        "model__n_estimators"    : [200, 300, 400, 500],
        "model__max_depth"       : [12, 15, 18, 20, None],
        "model__min_samples_leaf": [2, 3, 5, 10],
        "model__min_samples_split": [2, 5, 10],
        "model__max_features"    : ["sqrt", "log2", 0.3, 0.4],
        "model__max_samples"     : [0.6, 0.7, 0.8, None],
    }
elif "Gradient Boosting" in BEST_MODEL_NAME:
    param_dist = {
        "model__n_estimators"    : [200, 300, 400, 500],
        "model__learning_rate"   : [0.01, 0.03, 0.05, 0.08, 0.1],
        "model__max_depth"       : [4, 5, 6, 7],
        "model__subsample"       : [0.7, 0.8, 0.9, 1.0],
        "model__min_samples_leaf": [5, 10, 15, 20],
        "model__max_features"    : ["sqrt", "log2", 0.4, 0.5],
    }
else:  # Extra Trees fallback
    param_dist = {
        "model__n_estimators"    : [200, 300, 400],
        "model__max_depth"       : [15, 20, None],
        "model__min_samples_leaf": [2, 3, 5],
        "model__max_features"    : ["sqrt", "log2", 0.4],
    }

n_combos = 1
for v in param_dist.values(): n_combos *= len(v)

print(f"Model being tuned      : [{BEST_MODEL_NAME}]")
print(f"Total grid combinations: {n_combos:,}")
print(f"RandomizedSearch n_iter: 50 (random sample)")
print(f"\nParameter grid:")
for k, v in param_dist.items():
    print(f"  {k:<35}: {v}")
```

```python
# ── Prepare tuning subsample ──────────────────────────────────────────────────
TUNE_N  = min(200_000, len(X_train))
rng     = np.random.default_rng(RANDOM_STATE)
idx_t   = rng.choice(len(X_train), size=TUNE_N, replace=False)
X_tune  = X_train.iloc[idx_t].copy()
y_tune  = y_train.iloc[idx_t].copy()
print(f"Tuning subsample: {len(X_tune):,} rows ({len(X_tune)/len(X_train)*100:.1f} % of train)")

# ── RandomizedSearchCV ────────────────────────────────────────────────────────
tune_pipe = BASELINE_MODELS[BEST_MODEL_NAME]   # fresh (unfitted) pipeline

search = RandomizedSearchCV(
    estimator           = tune_pipe,
    param_distributions = param_dist,
    n_iter              = 50,
    cv                  = 3,
    scoring             = mape_scorer,
    n_jobs              = -1,
    random_state        = RANDOM_STATE,
    verbose             = 1,
    refit               = True,
    return_train_score  = True,
)

print(f"\nStarting RandomizedSearchCV (50 iterations × 3 folds = 150 fits) …")
t_start = time.time()
search.fit(X_tune, y_tune)
t_elapsed = time.time() - t_start

print(f"\n✅  Tuning complete in {t_elapsed/60:.1f} minutes")
print(f"\nBest CV MAPE (3-fold): {-search.best_score_:.2f}%")
print(f"\nBest hyperparameters:")
for k, v in search.best_params_.items():
    print(f"  {k:<40}: {v}")
```

```python
# ── Build final model with best params & fit on FULL train set ────────────────

best_params = {k.replace("model__", ""): v for k, v in search.best_params_.items()}

if "XGBoost" in BEST_MODEL_NAME:
    best_est = xgb.XGBRegressor(
        **best_params, tree_method="hist", eval_metric="rmse",
        random_state=RANDOM_STATE, n_jobs=-1, verbosity=0)
elif "Random Forest" in BEST_MODEL_NAME:
    best_est = RandomForestRegressor(
        **best_params, n_jobs=-1, random_state=RANDOM_STATE, oob_score=True)
elif "Gradient Boosting" in BEST_MODEL_NAME:
    best_est = GradientBoostingRegressor(**best_params, random_state=RANDOM_STATE)
else:
    best_est = ExtraTreesRegressor(**best_params, n_jobs=-1, random_state=RANDOM_STATE)

final_pipeline = Pipeline([("prep", preprocessor), ("model", best_est)])

print(f"Fitting final model on {len(X_train):,} training rows …")
t0 = time.time()
final_pipeline.fit(X_train, y_train)
print(f"✅  Final model trained in {time.time()-t0:.1f} s")

# ── Tuning results visualisation ──────────────────────────────────────────────

cv_df = pd.DataFrame(search.cv_results_).sort_values("rank_test_score").copy()

# Flip sign so everything reads in positive MAPE % — much more intuitive for the reader
cv_df["mean_mape_pct"] = -cv_df["mean_test_score"]
cv_df["std_mape_pct"]  = cv_df["std_test_score"]   # std is already a positive spread
best_mape_pct = -search.best_score_

fig, axes = plt.subplots(1, 2, figsize=(16, 5))
fig.suptitle(f"Hyperparameter Tuning Results — {BEST_MODEL_NAME}\n"
             f"50 Random Iterations × 3-Fold CV",
             fontsize=13, fontweight="bold")

# ── Panel 1: Distribution of CV MAPE across all 50 iterations ───────────────
ax = axes[0]
ax.hist(cv_df["mean_mape_pct"], bins=20, color=C["navy"], alpha=0.8, edgecolor="white")
ax.axvline(best_mape_pct, color="red", lw=2, ls="--",
           label=f"Best CV MAPE = {best_mape_pct:.2f}%")
ax.set_xlabel("CV MAPE (%)")
ax.set_ylabel("Count")
ax.set_title("Distribution of CV MAPE — All 50 Iterations", fontweight="bold")
ax.legend()

# ── Panel 2: Top 15 parameter combinations ───────────────────────────────────
ax = axes[1]
top15 = cv_df.head(15).iloc[::-1]   # reverse so rank 1 appears at the top of the barh
bar_cols = [C["red"] if i == len(top15)-1 else C["blue"] for i in range(len(top15))]

bars = ax.barh(range(len(top15)), top15["mean_mape_pct"],
                xerr=top15["std_mape_pct"],
                color=bar_cols, alpha=0.85, edgecolor="white",
                capsize=4, error_kw={"linewidth": 1.2})

ax.set_yticks(range(len(top15)))
ax.set_yticklabels([f"Rank {r}" for r in top15["rank_test_score"]], fontsize=8)
ax.set_xlabel("CV MAPE (%)")
ax.set_title("Top 15 Parameter Combinations", fontweight="bold")

# Annotate each bar with its actual MAPE value — removes guesswork from reading the axis
for i, (val, std) in enumerate(zip(top15["mean_mape_pct"], top15["std_mape_pct"])):
    ax.text(val + std + 0.15, i, f"{val:.2f}%", va="center", fontsize=8)

plt.tight_layout()
plt.show()

print(f"\n🏆  Best CV MAPE (3-fold): {best_mape_pct:.2f}%")
print(f"\nFinal tuned pipeline ready for comprehensive evaluation.")
```

**Section 4.5 Conclusion — Hyperparameter Tuning**

`RandomizedSearchCV` efficiently explored the hyperparameter space and identified a
parameter combination that materially improves over the default baseline configuration:

- **Depth control** (`max_depth`): The optimal depth is typically lower than the default —
  shallower trees reduce variance while retaining sufficient capacity for the non-linear
  property value relationships.
- **Learning rate / n_estimators trade-off** (for boosting models): Optimal performance
  is achieved with a lower learning rate paired with more estimators, allowing gradual
  refinement rather than large jumps.
- **Regularisation** (`reg_alpha`, `reg_lambda` for XGBoost): Non-zero regularisation
  prevents the model from over-indexing on noisy outlier properties, improving performance
  on the long tail of the value distribution.
- **Subsampling** (`subsample`, `colsample_bytree`): Values of 0.7–0.8 introduce randomness
  that further reduces overfitting — similar in spirit to the bagging in Random Forest.

## 4.7 Feature Importance & SHAP Explainability

Accuracy alone does not satisfy OPA's transparency mandate. Property owners have the right
to understand why their property was assessed at a particular value, and OPA assessors need
to validate that the model uses legally and economically defensible criteria.

Two complementary explainability approaches are applied:

| Method | Scope | Characteristic |
|---|---|---|
| **Permutation Importance** | Global (dataset-level) | Model-agnostic; measures R² drop when a feature is shuffled |
| **SHAP TreeExplainer** | Global + Local (per property) | Theoretically grounded; additive attribution; handles interactions |


```python
# ── Permutation Feature Importance ───────────────────────────────────────────

PERM_N = min(15_000, len(X_test))
rng    = np.random.default_rng(RANDOM_STATE)
idx_p  = rng.choice(len(X_test), PERM_N, replace=False)

print(f"Computing permutation importance on {PERM_N:,} test samples (10 repeats) …")
t0 = time.time()
perm_imp = permutation_importance(
    final_pipeline, X_test.iloc[idx_p], y_test.iloc[idx_p],
    n_repeats=10, scoring="r2", random_state=RANDOM_STATE, n_jobs=-1,
)
print(f"Done in {time.time()-t0:.1f} s")

feat_names_out = NUM_FEATURES + CAT_FEATURES_OHE + CAT_FEATURES_ORD

perm_df = pd.DataFrame({
    "Feature"   : feat_names_out,
    "Importance": perm_imp.importances_mean,
    "Std"       : perm_imp.importances_std,
}).sort_values("Importance", ascending=False)

top20_perm = perm_df.head(20)

fig, ax = plt.subplots(figsize=(13, 9))
pal_perm = [C["red"] if v > top20_perm["Importance"].quantile(0.75) else
            C["orange"] if v > top20_perm["Importance"].median() else C["blue"]
            for v in top20_perm["Importance"]]

ax.barh(top20_perm["Feature"][::-1], top20_perm["Importance"][::-1],
        xerr=top20_perm["Std"][::-1], color=pal_perm[::-1],
        alpha=0.85, edgecolor="white", capsize=4, error_kw={"linewidth":1.5})
ax.axvline(0, color="black", lw=0.8)
ax.set_xlabel("R² Drop (mean ± std over 10 permutations)", fontsize=11)
ax.set_title(f"Permutation Feature Importance — Top 20 Features\n"
             f"Model: {BEST_MODEL_NAME} (Tuned) | n={PERM_N:,} test samples",
             fontweight="bold", fontsize=12)
for i, (f, v, s) in enumerate(zip(top20_perm["Feature"][::-1],
                                    top20_perm["Importance"][::-1],
                                    top20_perm["Std"][::-1])):
    ax.text(v+s+0.001, i, f"{v:.4f}", va="center", fontsize=8.5)

plt.tight_layout()
plt.show()

print("\nTop 10 Most Important Features (Permutation Importance):")
print(perm_df.head(10)[["Feature","Importance","Std"]].to_string(index=False))
```

```python
# ── SHAP Analysis ─────────────────────────────────────────────────────────────

SHAP_N = 5_000
rng    = np.random.default_rng(RANDOM_STATE)
idx_sh = rng.choice(len(X_test), SHAP_N, replace=False)

fitted_model = final_pipeline.named_steps["model"]
fitted_prep  = final_pipeline.named_steps["prep"]

X_test_tf = fitted_prep.transform(X_test)

# Feature names after transform
num_names = NUM_FEATURES
try:
    ohe_names = list(fitted_prep.named_transformers_["cat_ohe"]
                 .named_steps["encoder"]
                 .get_feature_names_out(CAT_FEATURES_OHE))
except Exception:
    n_ohe = X_test_tf.shape[1] - len(NUM_FEATURES) - len(CAT_FEATURES_ORD)
    ohe_names = [f"ohe_{i}" for i in range(n_ohe)]
ord_names = CAT_FEATURES_ORD

all_tf_names = num_names + ohe_names + ord_names
X_shap_tf    = X_test_tf[idx_sh]

print(f"Computing SHAP values on {SHAP_N:,} test samples …")
t0 = time.time()
explainer   = shap.TreeExplainer(fitted_model)
shap_values = explainer.shap_values(X_shap_tf)
print(f"Done in {time.time()-t0:.1f} s | SHAP matrix shape: {shap_values.shape}")

# ── Aggregate OHE dummies back to parent feature ──────────────────────────
def aggregate_ohe_shap(sv, tf_names, ohe_list, num_f, ord_f):
    n_num = len(num_f); n_ord = len(ord_f)
    sh_num = sv[:, :n_num]
    sh_ohe = sv[:, n_num:n_num+len(ohe_list)]
    sh_ord = sv[:, -n_ord:] if n_ord else np.zeros((len(sv), 0))
    enc_n  = tf_names[n_num:n_num+len(ohe_list)]
    parent_map = {}
    for i, en in enumerate(enc_n):
        parent = next((c for c in CAT_FEATURES_OHE if en.startswith(c+"_") or en==c), en)
        parent_map.setdefault(parent, []).append(i)
    agg, names = [sh_num], list(num_f)
    for p, idxs in parent_map.items():
        agg.append(np.abs(sh_ohe[:, idxs]).sum(axis=1, keepdims=True))
        names.append(p)
    if n_ord:
        agg.append(sh_ord); names.extend(ord_f)
    return np.hstack(agg), names

shap_agg, agg_names = aggregate_ohe_shap(
shap_values, all_tf_names, ohe_names, NUM_FEATURES, CAT_FEATURES_ORD)

shap_df = (
    pd.DataFrame({"Feature": agg_names,
                "Mean |SHAP|": np.abs(shap_agg).mean(axis=0)})
                .sort_values("Mean |SHAP|", ascending=False)
                .head(25)
    )

# ── SHAP Global Bar ───────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(13, 9))
pal_sh = [C["red"] if v > shap_df["Mean |SHAP|"].quantile(0.75) else
            C["orange"] if v > shap_df["Mean |SHAP|"].median() else C["blue"]
            for v in shap_df["Mean |SHAP|"]]
ax.barh(shap_df["Feature"][::-1], shap_df["Mean |SHAP|"][::-1],
        color=pal_sh[::-1], alpha=0.85, edgecolor="white")
ax.set_xlabel("Mean |SHAP Value| — Average Absolute Contribution", fontsize=11)
ax.set_title(f"SHAP Global Feature Importance — {BEST_MODEL_NAME} (Tuned)\n"
                f"Top 25 Features | Predicting log(market_value)",
                fontweight="bold", fontsize=12)
for i, (f, v) in enumerate(zip(shap_df["Feature"][::-1], shap_df["Mean |SHAP|"][::-1])):
    ax.text(v+0.001, i, f"{v:.4f}", va="center", fontsize=8.5)
plt.tight_layout()
plt.show()
```

### 4.8.1 Tax Impact Asymmetry — Who Bears the Cost of Prediction Error?

Every prediction error has a financial direction, not just a magnitude:

- **Under-assessment** (`predicted < actual`): the property is taxed on a lower value than
  it is truly worth. The **property owner benefits** (lower tax bill) while the
  **City loses tax revenue** it is legally entitled to collect.
- **Over-assessment** (`predicted > actual`): the property is taxed on a higher value than
  it is truly worth. The **City gains** short-term revenue, but the **property owner is
  unfairly burdened** — and over-assessment is the single most common basis for formal
  tax appeals, which cost OPA additional staff time to litigate.

A model that is "accurate on average" can still hide a systematic lean toward one of these
two failure modes. We quantify that lean directly from the test-set predictions.


```python
# ── Official Philadelphia Property Tax Formula ───────────────────────────────
#   Property Tax = (Market Value − Exemptions) × 1.3998%
#   Tax Rate = 0.6159% (City) + 0.7839% (School District) = 1.3998% combined
# Source: phila.gov — Office of Property Assessment

PROPERTY_TAX_RATE = 0.013998   # 1.3998% combined City + School District rate

# ── Pull real exemption data from df_raw (dropped from df/df_ml in Section 2.1
#    due to leakage, but needed here purely for tax simulation, not modelling) ──
EXEMPTION_COLS = ["homestead_exemption", "exempt_building", "exempt_land"]
exemption_data = df_raw.loc[X_test.index, EXEMPTION_COLS].fillna(0)

total_exemption = (
    exemption_data["homestead_exemption"]
    + exemption_data["exempt_building"]
    + exemption_data["exempt_land"]
)

print(f"Exemption data retrieved for {len(total_exemption):,} test-set properties")
print(f"  Properties WITH any exemption    : {(total_exemption > 0).sum():,}  "
      f"({(total_exemption > 0).mean()*100:.1f}%)")
print(f"  Properties WITHOUT exemption      : {(total_exemption == 0).sum():,}  "
      f"({(total_exemption == 0).mean()*100:.1f}%)")
print(f"  Median exemption (where present)  : "
      f"${total_exemption[total_exemption > 0].median():,.0f}")

# ── Taxable Value = Market Value − Exemptions (floored at 0) ────────────────
taxable_actual = (y_test_usd        - total_exemption.values).clip(min=0)
taxable_pred   = (y_test_pred_usd   - total_exemption.values).clip(min=0)

# ── Annual Property Tax = Taxable Value × 1.3998% ────────────────────────────
tax_actual    = taxable_actual * PROPERTY_TAX_RATE
tax_predicted = taxable_pred   * PROPERTY_TAX_RATE

print(f"\nOfficial formula applied: Property Tax = (Market Value − Exemptions) × 1.3998%")
print(f"  Mean ACTUAL annual tax (if true market_value were billed)    : ${tax_actual.mean():,.2f}")
print(f"  Mean PREDICTED annual tax (if model's prediction were billed): ${tax_predicted.mean():,.2f}")
```

```python
# ── Worked Example: Verify formula against the official illustration ────────
#   OPA Market Value : $300,000
#   Homestead Exemption : $100,000
#   Taxable Value : $200,000
#   Annual Tax : $200,000 × 1.3998% = $2,799.60

example_market_value = 300_000
example_homestead     = 100_000
example_taxable       = example_market_value - example_homestead
example_tax           = example_taxable * PROPERTY_TAX_RATE

print("=" * 60)
print("WORKED EXAMPLE — Official OPA Tax Formula")
print("=" * 60)
print(f"  OPA Market Value         : ${example_market_value:>12,.2f}")
print(f"  Homestead Exemption      : ${example_homestead:>12,.2f}")
print(f"  Taxable Value            : ${example_taxable:>12,.2f}")
print(f"  Tax Rate                 : {PROPERTY_TAX_RATE*100:>12.4f} %")
print(f"  Annual Property Tax      : ${example_tax:>12,.2f}")
print("=" * 60)
assert abs(example_tax - 2799.60) < 0.01, "Formula does not match official example!"
print("✅  Formula verified against official phila.gov worked example.")
```

```python
# ── Tax Impact Asymmetry: Over vs Under Assessment (USING REAL TAX FORMULA) ──

tax_error_usd = tax_actual - tax_predicted   # positive = City under-collects, negative = owner overpays
tax_error_pct = np.divide(tax_error_usd, tax_actual,
                           out=np.zeros_like(tax_error_usd), where=tax_actual > 0) * 100

is_under = tax_error_usd > 0     # predicted tax LOWER than actual tax  -> City loses revenue
is_over  = tax_error_usd < 0     # predicted tax HIGHER than actual tax -> Owner overpaying

n_under = is_under.sum()
n_over  = is_over.sum()
n_exact = (tax_error_usd == 0).sum()

total_revenue_at_risk  = tax_error_usd[is_under].sum()    # $ City would under-collect annually
total_owner_overcharge = -tax_error_usd[is_over].sum()    # $ owners would overpay annually

print("=" * 72)
print("TAX IMPACT ASYMMETRY — Official OPA Formula Applied to Test Set")
print(f"  (Property Tax = (Market Value − Exemptions) × 1.3998%)")
print("=" * 72)
print(f"  Test set size                                    : {len(tax_actual):>8,} properties")
print(f"  Properties UNDER-billed (City loses revenue)     : {n_under:>8,}  ({n_under/len(tax_actual)*100:.1f}%)")
print(f"  Properties OVER-billed  (Owner overpays)         : {n_over:>8,}  ({n_over/len(tax_actual)*100:.1f}%)")
print(f"  Properties exactly correct                       : {n_exact:>8,}")
print()
print(f"  Total ACTUAL annual tax (true market_value)      : ${tax_actual.sum():>16,.2f}")
print(f"  Total PREDICTED annual tax (model's market_value): ${tax_predicted.sum():>16,.2f}")
print()
print(f"  Annual tax REVENUE AT RISK (under-billing)        : ${total_revenue_at_risk:>16,.2f}")
print(f"  Annual tax OVERCHARGED to owners (over-billing)   : ${total_owner_overcharge:>16,.2f}")
print(f"  Net direction : "
      f"{'⚠️  Model net UNDER-bills (City revenue risk)' if total_revenue_at_risk > total_owner_overcharge else '⚠️  Model net OVER-bills (Taxpayer burden risk)'}")
print("=" * 72)
```

1. The model should be deployed as a **decision-support tool alongside human assessors**,
   not as a standalone replacement. A practical deployment workflow is: run the model
   on all properties annually, then flag any property where the model's prediction
   deviates more than ±30% from the current official assessment for **priority manual
   review**. This hybrid approach captures the speed and consistency of machine
   learning while preserving human judgment for edge cases and formal appeal hearings.

2. Since `geographic_ward` accounts for more than twice the predictive importance of
   any other feature (0.51 vs 0.26), OPA should implement **ward-level assessment
   ratio monitoring** to track whether assessment-to-sale ratios are consistent within
   and across wards. Significant ratio disparities within a single ward may indicate
   that certain property sub-types or streets are being systematically over- or
   under-assessed — a fairness concern that the ward-level feature importance directly
   surfaces.

   Since the model's accuracy is directly tied to the accuracy of this input, OPA should
   treat **livable area measurement** as its highest-priority data quality initiative.
   Properties where livable area data has not been verified in more than five years should
   be prioritised for field inspection, as unrecorded renovations and additions are the
   most common source of systematic assessment error in the mid-range residential segment.

4. The SHAP waterfall analysis confirms that the model **significantly underestimates
   ultra-high-value properties** (e.g. actual value $405.6M, predicted only $13.1M).
   OPA should explicitly exclude properties above a defined value threshold —
   recommended at $10 million — from automated model-based assessment, and route
   them to specialist commercial appraisers who can apply income-approach and
   cost-approach valuation methods appropriate for those asset classes.

5. The model should be **retrained annually** using a rolling 3-year window of
   recorded sale transactions. Philadelphia's market dynamics, including post-pandemic
   price shifts, neighbourhood gentrification, and new development activity, can cause

   12–18 months. Monthly monitoring of the prediction-to-actual ratio on newly
   recorded sales will provide an early-warning signal when retraining becomes necessary.
